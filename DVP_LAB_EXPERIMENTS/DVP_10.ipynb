```python
# ==============================================================
#   Ex No: 10  Implementation of SVM Classification Techniques
# ==============================================================

import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# --------------------------------------------------------------
# 1. Load Breast Cancer Wisconsin Dataset
# --------------------------------------------------------------

cancer = load_breast_cancer()

# Convert to DataFrame for easy viewing
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['target'] = cancer.target

print("Dataset Loaded Successfully!\n")
print(df.head())

# --------------------------------------------------------------
# 2. Split dataset into Train and Test
# --------------------------------------------------------------

X = df.drop('target', axis=1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print("\nTraining samples:", len(X_train))
print("Testing samples:", len(X_test))

# --------------------------------------------------------------
# 3. Train SVM Classifier
# --------------------------------------------------------------

svm_model = SVC(kernel='linear')   # Using linear SVM
svm_model.fit(X_train, y_train)

print("\nSVM Model Training Completed!")

# --------------------------------------------------------------
# 4. Predict on Test Data
# --------------------------------------------------------------

y_pred = svm_model.predict(X_test)

# --------------------------------------------------------------
# 5. Evaluate Accuracy
# --------------------------------------------------------------

accuracy = accuracy_score(y_test, y_pred)

print("\n==============================")
print("SVM Classification Evaluation")
print("==============================")
print("Accuracy:", accuracy)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# --------------------------------------------------------------
# 6. Plot Correlation Heatmap (Optional Visualization)
# --------------------------------------------------------------

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```
